```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
datos = pd.read_csv("Salary_Data.csv")
```

```python
datos
```

```python
X = datos[["YearsExperience"]]
```

```python
X
```

```python
y= datos["Salary"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.7 , random_state=0)
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.score(X_test,y_test)
```

```python
regression.predict(X_test,)
```

```python
plt.scatter(datos["YearsExperience"],datos["Salary"])
plt.plot(X_train, regression.predict(X_train))
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
logistic.score(X_train,y_train)
```

```python
print(logistic.predict(X_train))
print(regression.predict(X_train))
```

```python
plt.scatter(datos["YearsExperience"],datos["Salary"])
plt.plot(datos["YearsExperience"], logistic.predict(X), color="red")
```
